```python
#now we need yo extract only the name of the gnere part so we will fomr a function in loop using the fnction def 
#lets take convert fumction as the helper function
#forming loop from i in object .
#here object in short form obj is our genre column function.
#we need to extract only the name from the object.
#problem here is the our object content of genre is in the form if the strings whih we first need to convert it into list so that we can append the list only for the name part.

```

```python
import ast 
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#func to correct format
def convert(obj):
    import ast
    L=[]
    for i in ast.literal_eval(obj): #converted to list
        L.append(i['name']) #format corrected
    return L
```

```python
 #here only for the genre part we apply convert helper function used in def function to form the correct format
 movies['genres'] = movies['genres'].apply(convert)
```

```python
#same we will do for the keywords columns by applying the convert function as it also holds the category of name.
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['cast'][0] #as you can see form here that their are several info about the cast team and we need to extract only the names of the starting three main charcters.
```

```python
#in the cast column we only need to focus on the 3 main characters real name of the movie
def convert_cast(obj):
    import ast
    L=[]
    counter= 0
    for i in ast.literal_eval(obj): #converted to list
      if counter!= 3:
        L.append(i['name']) #format corrected
        counter+=1
      else:
        break
    return L
```

```python
movies['cast']= movies['cast'].apply(convert)
```

```python
movies['cast'].head()
```

```python
movies['crew'].values
```

```python
#from the above text we only need to focus on the director word from the label of job.
#so defining a function to make the column crew in a good format.
def director(obj):
    import ast
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']== 'Director':
            L.append(i['job'])
            break
    return L
    
```

```python
movies['crew']=movies['crew'].apply(director)
```

```python
movies['crew'].head()
```

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
#NEXT STEP 
#Now all the coloum contents are in the form of lists so that we can merge them easily.
#Here one big problem arises as the names are separated with spaces between them.
#for eg. if there are two diff guys sam with diff surnames then if we dont combine their names then the system will
# consider both the names and surnames as the two different words or names.
#in the dataset their are two actors as the same name as 'sam' (sam warthington and sam mendes)
#but they both are from different due to their surnames so we need to join their names with their surnames to make sure that there is no problem created 
```

```python
#syntax for joining two words together(here name and surname)
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x]) # i.replace = replace the space(" ") with nothing("") for every x.
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#combining the keyword,cast,crew,overview and genre in a single coloumn names tag.
movies['tags']=movies['keywords']+movies['cast']+movies['crew']+movies['genres']+movies['overview']
```

```python
#forming a new dataframe as we received 3 new columns named as id,title,tags.

new_df= movies[['id','title','tags']]
```

```python
#Major step:
#here we will create the vectors of the tags of the movies.
#by creating vectors we can compare the movies on how closely they are related ot each other.
#Process of creating vectors of tags: step 1> Firstly we will combine all the tags of the movies in a single tag.(lessay suppose each movie tag has 20 words so in toal there will be 20+20+...upto the no of movies words)
#step 2> we will make a list of total 5000 words in our case and now we will make a table which consists of the no of words in different column and no of movies in different rows.
#step 3> we will now find out the frequency that how many times a particular word got repeated in a single movie.(first we will find for movie 1 all the words which are present in the movie and then for movie2)
#step 4>if the frequency of different words between 2 movies matches then or are nearer then we can say that the 2 movies are relatable to each other and we can recommend them.

```

```python
#here a big problem arises before comparing 2 words that some words like (actions,action) and (loving,loved) are same only but in comparing them the result will be different.
#also we cannot compare words like a,in,as,the in the word frame as they all are considered as stop words in python,so we need to remove them as well.
#making a unique form for the similar words like (actions,action) by using library nltk and the stem class and porterstemer sub class or function inside the nltk.
import nltk #importing nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
#defining a function named as stem.
def stem(text):
    y=[]
    for i in text.split(): #making the strings as list
        y.append(ps.stem(i)) #this is list obtained by only having unique words for similar word kind.
        
    return " ".join(y) 
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
#Feature Extraction :
#the mapping from textual data to real valued vectors is called as feature extraction.
# the usage of this feature extraction came into the scenario because the ml learning algo find it hard to analyze the text data as compared to numerical data.
#thats why we use this feature.
```

```python
#Countvectorization refers a the breaking down of the sentence or texts into words by performing various preprocessing task.
#moving onto the making vectors of the movies using countvectorizer using library scikit-learn
from sklearn.feature_extraction.text import CountVectorizer #sklearn is scikit learn
cv= CountVectorizer(max_features=5000,stop_words='english') #here max feature is the extracted words while all tags are combined together and stop words are the unuseful english words like as,in,the,a..etc

vector=cv.fit_transform(new_df['tags']).toarray() #converting to array format using fit_transform feature
```

```python
#when all the movies are presented in the vector format they will be compared with each other by calculating the cosine distance between them.
#their is class of scikit learn library with which we can make use of cosine distance(lesser the cosine distance more closely two movies will be associated to each other
#
```

```python
from sklearn.metrics.pairwise import cosine_similarity  #cosine_similarity calculates cosine distance between two functions(here movies)
similarity =cosine_similarity(vector)   #putting vectors of movies in cosine similarity to calculate how similar they are to each other,the more similar hey are the more value they will pursue as cos0 =1.(which shows less angle between them)
similarity[0] #it shows how the 1st movie with index 0(avatar) is similar to the other 4806 movies.

#in the array shown the movie with index 0 when its similarity is calculated with the movie with index 0,index 1,index 2......index 4805(as their are total 4806 movies) will be:
```

```python
sorted(similarity[0],reverse=True) #sorting is performed in ascending order but we need to find those who are in descending order,so applying reverse function.
# list of most similar movies with the 1st movie with index 0 i.e. avatar is obtained we need to extract only the 5 most similar movies.
# since after sorting the values for the 0th index movie avatar the index of them are also sorted respectively as the similarity[0]
#used to give the similarity of the 1st movie with 1st movie,1st movie with the 2nd movie,1st movie with the 3rd movie etc.
#the main problem is that we need to tell that the phle movie ka kis index ki movie ke saath kitna similairty hai,yaha pe similairty ke basis pe sorting hui hai usse puri movies ka index uthal puthal hogya isliye humko indexes ki position same rakhni hogi baaki sorting sirf similarities ke baisi pe krna hoga.
#so we need to preserve the indexes of the movies by using enumerate function.
```

```python
sorted(list(enumerate(similarity[0])),reverse=True) #enumerate function applied to similarity[0] will not lead similarity to loose the indexes of the movies.
#here although the sorting is performed but on the basis of the indexes of the movies and not on the basis of similarity,so nwe will use key and lambda function for this issue.
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1]) #x[1] means the second position of the list i.e. the similarities calculates and not the index of the movies.
```

```python
#defining a function to recommend the movies.
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0] #from eq 1
    distance=similarity[movie_index]#here just like for avatar similarity[0] '0'gives the index of the movie avatar just like if we need to calculate the index similarity for other movies with the rest of the movies we need to pass on the index of them in the similarity function.
    #now to find out the movies which are similar and can be recommended we need to find out the 5 most match founding movies with a particular movies of which we need to find the ans.
    #for this we will do sorting which is given above.
    
    movies_list= sorted(list(enumerate(distance)), reverse= True, key= lambda x:x[1])[1:6] #here the [1,6] represents the first 5 movies and for every movie we put distance.
    for i in movies_list:
       print(new_df.iloc[i[0]].title) #i[0] represents the 1st tuple of the list.
    
```

```python
recommend('Avatar') #will recommend similar movies.
```
